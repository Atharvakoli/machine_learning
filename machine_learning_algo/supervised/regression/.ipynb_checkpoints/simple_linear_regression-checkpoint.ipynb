```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline # visualization should be their within this jupyter notebook
```

```python
df = pd.read_csv('../datasets/height-weight.csv')
```

```python
df.head()
```

```python
## scatter plot
plt.scatter(df['Weight'], df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
## Correlation
df.corr()
```

```python
## Seaborn for visualization
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
## Independent and dependent features
# Independent features
# X_series = df['Weight']
# print(np.array(X_series).shape)

# Why do we say this as regression problem statement because output variable is continuous

X = df[['Weight']] ## Independent features hsould be dataframe or 2 dimensions array np.array(X)

# Dependent Features

y  = df['Weight'] ## dependent or Output feature ## this variable can be in series or 1d array
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
```

```python
# Train => To train the model
# Test => use for Testing the model
X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() ## this will initialize that formula z-score = (data point - meu) / standard deviation
X_train = scaler.fit_transform(X_train)
# When I say transform, that basically means I am going to use the training data set mean and standard deviation and use it over Test dataset
# When I say fit_tranform then again mean or standard deviation will get calculated for this data
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
## Apply Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression(n_jobs=-1) ## n_jobs runs all processor on system
```

```python
regression.fit(X_train, y_train)
```

```python
regression
```

```python
print("Coefficent or slope", regression.coef_) ## y hate = beta0 (Intercept) + beta1 (16.50710238) * X1
print("Intercept: ", regression.intercept_)
```

```python
## plot Training data plot bestfit line
plt.scatter(X_train,y_train)
plt.plot(X_train, regression.predict(X_train))
plt.show()
```

## Prediction of test data
1. predicted height output = intercept + cof_(weights)
2. y_pred_test = 72.470 + 16.50(X_test)

```python
## Prediction for test data
Y_pred = regression.predict(X_test)
```

```python
## Performance Matrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, Y_pred)
mae = mean_absolute_error(y_test,Y_pred)
rmse = np.sqrt(mse)
print("Mean Squared error: ", mse,"\n", 
      "Mean Absolute error", mae, "\n", 
      "Root Mean squared error", rmse)
```

## R square formula

R^2 (coefficient of determination) = 1 - SSR (sum of square of residuals) / SST (total sum of squares)


```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, Y_pred)
print(score)
```

## Adjusted R2 = 1 - [(1 - R2) * (n-1) / (n - k - 1)]
Where:
1. R2: The R2 of the model
2. n: The number of observations
3. k: The number of predictor variables

```python
## display adjusted R-squared
1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
```
