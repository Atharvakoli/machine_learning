### Combining datasets

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
# Tocheck datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## pie chart - Top 3 countries that used zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

Observation: Zomato maximum recordes or transactions are from India After that USA and then United kingdom

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

Observation
 1. When Rating is between 4.5 to 4.9  --> Ecellent
 2. When ratings are between 3.5 to 3.9 --> Very Good
 3. When Rating is between 3.5 to 3.9 --> good
 4. When Rating is between 3.0 to 3.4 --> Average
 5. When Rating is between 2.5 to 2.9 --> Average
 6. When Rating is between 2.0 to 2.4 -->poor

```python
rating.head()
```

```python
mpl.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=rating)
plt.show()
```

```python
mpl.rcParams['figure.figsize'] = (12,6)
sns.barplot(
    x="Aggregate rating", 
    y="Rating Count", 
    hue='Rating color', 
    data=rating, 
    palette=['olive', 'red', 'orange', 'yellow', 'green', 'green']
)
plt.show()
```

Observation: 
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(
    x="Rating color", 
    data=rating, 
    palette=['olive', 'red', 'orange', 'yellow', 'green', 'green'],
    legend=False
)
plt.show()
```
