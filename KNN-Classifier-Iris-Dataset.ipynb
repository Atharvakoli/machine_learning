# KNN Classifier

- Iris Dataset

### Step 1: Import necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the dataset (Iris dataset in this case)

```python
data = pd.read_csv('datasets/loan.csv')
data
```

```python
data.head()
```

```python
data.shape
```

```python
data = data[['term', 'int_rate', 'grade']]
data
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoders
term_encoder = LabelEncoder()
grade_encoder = LabelEncoder()

# Encode 'term' and 'grade'
data['term'] = term_encoder.fit_transform(data['term'])
data['grade'] = grade_encoder.fit_transform(data['grade'])

# Drop original categorical columns if needed
df_encoded = data.drop(['term', 'grade'], axis=1)

# Final result
print(df_encoded)
```

```python
data
```

### Step 3: Data Preprocessing

```python
X = data.drop('grade', axis=1)  # Features
y = data['grade']  # Target variable
```

### Step 4: Split the dataset into a training set and a testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Function to choose the value of k
def choose_k(X_train, y_train, X_test, y_test, max_k=10):
    k_values = list(range(1, max_k+1))
    accuracy_values = []
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_values.append(accuracy)
    
    # Visualize the accuracy for different k values
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, accuracy_values, marker='o', linestyle='-')
    plt.title('Accuracy vs. Number of Neighbors (k)')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Accuracy')
    plt.xticks(k_values)
    plt.grid(True)
    plt.show()

    best_k = k_values[accuracy_values.index(max(accuracy_values))]
    return best_k
```

```python
# Choose the best k value
best_k = choose_k(X_train, y_train, X_test, y_test)
print("Best k:", best_k)
```

### Step 5: Create and train the KNN model with the best k

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)
```

### Step 6: Make predictions

```python
y_pred = knn.predict(X_test)
```

### Step 7: Evaluate the model

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
